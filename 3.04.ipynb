```python
from statistics import linear_regression
from sklearn.inspection import permutation_importance
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
import pandas as pd
import matplotlib.pyplot as plt
import math
import sklearn
from sklearn.linear_model import LinearRegression as ls
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler, PowerTransformer, QuantileTransformer
import seaborn as sns
from sklearn.model_selection import train_test_split


from analysis import Analysis
```

```python
#data explanation
analyzer = Analysis()
data = analyzer.loadData('data/raw_data/zal_hausing.csv')
print(data.head().to_string())
analyzer.findMissingValues(data)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
print(y)
knn.fit(X,y)
predictions = knn.predict(X)
print(predictions)
```

```python
scaler = StandardScaler().fit(data)
scaled = scaler.transform(data)
X = scaled[:, :-1]
y = data.iloc[:, -1]

X_train,X_test,y_train,y_test = train_test_split(X,y)
estimator = svm.SVC()
estimator.fit(X_train,y_train)
y_pred = estimator.predict(X_test)

trained = permutation_importance(estimator, X_train, y_train, n_repeats=30,random_state=0)
tested = permutation_importance(estimator, X_test, y_test, n_repeats=30,random_state=0)


```

```python
pd.DataFrame(zip(trained['importances_mean'],tested['importances_mean']), columns=["train","test"])
```
